```python
@np.vectorize
def Step_exp(t,tp,Fp,b ,alpha):
    if t < tp:
        return float(0)  # uniform data types 
    if t >= tp:
        return float(Fp*(t/tp-b)**(-alpha))

x,y = np.genfromtxt('DATA/LC/Exponential/LAT_090926.txt', unpack=True, skip_header=1 ) ; y=y-y[0]


plt.plot(x,y,'-',color='grey', label='$Fermi$-measured LC')
x2 =np.linspace(-300,300,1000)
y2 = Step_exp(x2,0.71,0.27,-3,0.95)/ Step_exp(0.71,0.71,0.27,-3,0.95)
plt.plot(x2,y2, color=tugreen, label='Fitted exponential decay')

plt.legend()
plt.ylabel(r'Normalized $\frac{\mathrm{Counts}}{\mathrm{Exposure}}$ / $\frac{1}{cm¬≤s}$')
plt.xlabel('Time / seconds after trigger')
plt.savefig('DPG/Exponential.jpg')
```

```python
def plot_Sens(scale,style):  ## Scale = string: 'Kev','Mev', 'Gev', 'Tev'              &  ## style = String: 'curve' oder 'binwise '
    Emin,Emax,S = np.genfromtxt('Sensitivity/CTA-Performance-prod3b-v1-South-20deg-30m-DiffSens.txt', unpack=True,skip_header=10)
    S_J = 1e-7*S
    S_ev = S_J/(elem)
    if scale == 'Kev':
        S = S_ev*1e-3
        Emax *=1e9
        Emin *=1e9
    if scale == 'Gev':
        S = S_ev*1e-9
        Emax *= 1e3
        Emin *= 1e3
    if scale == 'Tev':
        S = S_ev*1e-12
    if scale == 'Mev':
        S = S_ev*1e-6
        Emax *= 1e6
        Emin *= 1e6
    if style == 'curve':
        E_center = (Emax+Emin)/2
        plt.plot(E_center,S,'-', color='k',label='CTA Sensitivity - 30 min')
    if style == 'binwise' :
        E_center = (Emax+Emin)/2
        E_width = E_center-Emin
        plt.errorbar(E_center, S, xerr=E_width, yerr=None,fmt=None ,ecolor = 'k',label='CTA Sensitivity - 30 min')

    plt.xscale('log')
    plt.yscale('log')
    
    

def ctools_Crab(E,SED):
    index = 2.48
    C = 5.7e-19 / (u.cm**2 * u.s * u.keV)  ## e-16 in MeV
    E_0 = (0.3e6 * u.MeV).to(u.keV)
    factor = 1
    if SED == True:
        factor = E*E
    return C*(E/E_0)**(-index)*factor

def HEGRA_Crab(E,SED):
        # Spectrum HEGRA (2004)
        # index = 2.62
        # C = 2.83e-11 / (u.cm**2 * u.s * u.TeV)
        # E_0 = 1 * u.TeV
    index = 2.62
    C = 2.83e-20 / (u.cm**2 * u.s * u.keV)  ## e-16 in MeV
    E_0 = (1e6 * u.MeV).to(u.keV)
    factor = 1
    if SED == True:
        factor = E*E
    return C*(E/E_0)**(-index)*factor


def plot_ctools_Crab(SED):
    e_kev  =np.logspace(6,10)
    plt.plot(e_kev,ctools_Crab(e_kev,SED),color=tugreen, label='ctools Crab spectrum')
    plt.plot(e_kev,ctools_Crab(e_kev,SED)*0.1, color=tugreen, linestyle = '--', label='10 % ctools Crab spectrum')
    plt.plot(e_kev,ctools_Crab(e_kev,SED)*10, color=tugreen, linestyle = ':', label='1000 % ctools Crab spectrum')
    plt.xscale('log') ; plt.yscale('log') ; plt.legend()
    plt.ylabel(string) ; plt.xlabel('Energy / keV')
    
def plot_HEGRA_Crab(SED):
    e_kev  =np.logspace(6,10)
    plt.plot(e_kev,HEGRA_Crab(e_kev,SED),color=tugreen, label='HEGRA Crab spectrum')
    plt.plot(e_kev,HEGRA_Crab(e_kev,SED)*0.1,color=tugreen,linestyle='--', label='10 % HEGRA Crab')
    plt.plot(e_kev,HEGRA_Crab(e_kev,SED)*10,color=tugreen,linestyle=':', label='1000% HEGRA Crab')
    plt.xscale('log') ; plt.yscale('log') ; plt.legend()
    
    
# DataFrame from results of the  first Fermi LAT GRB Catalog (Paper, 2013 [arXiv:1303.2908] )
GRB_EXT = pd.read_csv('Kataloge/GRBs_Joint_Interval.csv',sep=' ',decimal=',')
def get_indices_from_LAT(GRB_name):
    GRB = GRB_EXT[GRB_EXT['Name'].str.contains(GRB_name)]
    index = GRB.iloc[0][0]
    Delta_t = ufloat(GRB['Time'][index],0) ## Time between LAT T05 and GBM T95
    BF = GRB['Best_Fit'][index]    ## Best Fit
    MFluence = ufloat(GRB['Main_Fluence'][index],GRB['Main_Fluence_err'][index]) ;
    AFluence = ufloat(GRB['Add_Fluence'][index],GRB['Add_Fluence_err'][index])    ## Fluence bewtween 10 keV and 10 GeV
    ## Band
    BE0 = ufloat(GRB['Band_E0'][index],GRB['Band_E0_err'][index]) ; BA = ufloat(GRB['Band_alpha'][index],GRB['Band_alpha_err'][index]) ;
    BB = ufloat(GRB['Band_beta'][index],GRB['Band_beta_err'][index])
    ## Compton
    CE0 = ufloat(GRB['Compton_E0'][index],GRB['Compton_E0_err'][index]) ; CA = ufloat(GRB['Compton_alpha'][index],GRB['Compton_alpha_err'][index])
    ## Log
    LB = ufloat(GRB['Log_b'][index],GRB['Log_b_err'][index]); LEP = ufloat(GRB['Log_Ep'][index],GRB['Log_Ep_err'][index])
    ## Additional +
    Plaw = ufloat(GRB['Plaw_alpha'][index],GRB['Plaw_alpha_err'][index]) ;  CUT = ufloat(GRB['Plaw_Cut'][index],GRB['Plaw_Cut_err'][index])
    return Delta_t,BF,MFluence,AFluence,BE0,BA,BB,CE0,CA,LB,LEP,Plaw,CUT

## Plot Differential Flux in log-log to energy with best fitting model or if SED = True, Diff. Flux * E¬≤
def plot_Flux_Energy_LAT(GRB_name,loglow,loghigh,SED,EBL,Redshift):  ## SED = Boolean, EBL = Boolean, then redshift needed
    Delt,BF,MFluence,AFluence,BE0,BA,BB,CE0,CA,LB,LEP,Plaw,CUT  = get_indices_from_LAT(GRB_name)
    E_lines = np.logspace(loglow,loghigh) ## keV
    Fitpoints = unp.uarray(np.zeros(len(E_lines)),np.zeros(len(E_lines)))
    Factor = 1
    string = r'Diff. Flux  / $\frac{\gamma}{kev cm¬≤s}$'
    style = '-'
    EBL_note = ''
    if SED == True:
        Factor = E_lines**2
        string = r'Flux $\cdot$ E¬≤ / $\frac{\gamma \mathrm{kev}}{\mathrm{cm}¬≤s}$'
    if EBL == True:
        Tau = np.zeros(len(E_lines))
        Tau = get_absorpt(Redshift,E_lines)
        Factor = Factor*Tau
        style = '--'
        EBL_note = '_EBL'

    if BF == 'Band':
        Int = quad(Bandfunc_LAT,10,10*1e6,args=(1,BA.n,BB.n,BE0.n,True)) ; Integral = ufloat(Int[0],Int[1])
        MFl = MFluence/(Delt*Integral)
        Fitpoints =  Bandfunc_LAT(E_lines,MFl,BA,BB,BE0,False)*Factor
        y = unp.nominal_values(Fitpoints) ; yerr = unp.std_devs(Fitpoints)
        plt.plot(E_lines,y, ls = style,linewidth = 0.8,color='grey', )#label='LAT_%s%s'%(GRB_name,EBL_note))
        #plt.fill_between(E_lines,abs(y-yerr),abs(y+yerr),color='grey',alpha=0.5)
    if BF == 'Band+Plaw':
        Int = quad(Bandfunc_LAT,10,10*1e6,args=(1,BA.n,BB.n,BE0.n,True)) ; Integral = ufloat(Int[0],Int[1])
        MFl = MFluence/(Delt*Integral)
        Intl = quad(Plaw_LAT,10,10*1e6,args=(1,Plaw.n,True)) ; Integral = ufloat(Int[0],Int[1])
        AFl = AFluence/(Delt*Integral)
        Fitpoints = Bandfunc_LAT(E_lines,MFl,BA,BB,BE0,False)*Factor+Plaw_LAT(E_lines,AFl,Plaw,False)*Factor
        y = unp.nominal_values(Fitpoints) ; yerr = unp.std_devs(Fitpoints)
        plt.plot(E_lines,y, ls = style,linewidth = 0.8,color='grey', )#label='LAT_%s%s'%(GRB_name,EBL_note))
        #plt.fill_between(E_lines,abs(y-yerr),abs(y+yerr),color='crimson',alpha=0.5)
    if BF == 'Band+Plaw*Cut':
        Int = quad(Bandfunc_LAT,10,10*1e6,args=(1,BA.n,BB.n,BE0.n,True))  ; Integral = ufloat(Int[0],Int[1])
        MFl = MFluence/(Delt*Integral)
        Int = quad(Plaw_LAT,10,10*1e6,args=(1,Plaw.n,True)) ; Integral = ufloat(Int[0],Int[1])
        AFl = AFluence/(Delt*Integral)
        Fitpoints =  Bandfunc_LAT(E_lines,MFl,BA,BB,BE0,False)*Factor+Plaw_CUT(E_lines,AFl,Plaw,CUT)*Factor
        y = unp.nominal_values(Fitpoints) ; yerr = unp.std_devs(Fitpoints)
        plt.plot(E_lines,y, ls = style,linewidth = 0.8,color='grey', label='Extrapolated $Fermi$ spectrum%s'%(EBL_note))
        #plt.fill_between(E_lines,abs(y-yerr),abs(y+yerr),color='grey',alpha=0.5)
    if BF == 'LogParabola':
        Int = quad(LogParabola_LAT,10,10*1e6,args=(1,LEP.n,LB.n,True),limit=100000) ; Integral = ufloat(Int[0],Int[1])
        MFl = MFluence/(Delt*Integral)
        Fitpoints =LogParabola_LAT(E_lines,MFl,LEP, LB,False)*Factor
        y = unp.nominal_values(Fitpoints) ; yerr = unp.std_devs(Fitpoints)
        plt.plot(E_lines,y, ls = style,linewidth = 0.8,color='grey',)#label='LAT_%s%s'%(GRB_name,EBL_note))
        #plt.fill_between(E_lines,abs(y-yerr),abs(y+yerr),color='crimson',alpha=0.5)

    plt.xscale('log'),plt.yscale('log')
    plt.legend(), plt.xlabel('Energy  / keV'), plt.ylabel(string)
    




```

```python
plot_Flux_Energy_LAT('GRB091003',3,10,True,False,1.99)
plot_Flux_Energy_LAT('GRB090926',3,10,True,False,1.99)
plot_Flux_Energy_LAT('GRB110721',3,10,True,False,1.99)
plot_Sens('Kev','binwise')
plot_HEGRA_Crab(True)
plt.savefig('DPG/SpectrumCrab.jpg')
```
