```python
from itertools import chain

def draw_map(m, scale=0.2):
    # draw a shaded-relief image
    m.shadedrelief(scale=scale)
    # lats and longs are returned as a dictionary
    lats = m.drawparallels(np.linspace(-90, 90, 6),labels=[1,0,0,0])
    lons = m.drawmeridians(np.linspace(-180, 180, 6),labels=[0,0,0,1])
    lat_lines = chain(*(tup[1][0] for tup in lats.items()))
    lon_lines = chain(*(tup[1][0] for tup in lons.items()))
    all_lines = chain(lat_lines, lon_lines)
    
    # cycle through these lines and set the desired style
    for line in all_lines:
        line.set(linestyle='-', alpha=0.3, color='w')

fig = plt.figure(figsize=(8, 6), edgecolor='w')
m = Basemap(projection='cyl', resolution=None,
            llcrnrlat=-90, urcrnrlat=90,
            llcrnrlon=-180, urcrnrlon=180, )
draw_map(m)

mlin = np.linspace(-180,180)
m.plot(mlin,0*mlin,'--', color='tan')

m.plot(-17.89,28.76,'o',color='tan', ms=12,label='CTA North')
m.plot(Angle('-70°24′15″')/u.deg,Angle('-24°37′38″')/u.deg,'s',color='tan',ms=12, label='CTA South')
m.plot(df_Optical['Longitude_East'],df_Optical['Latitude_North'],'*',color='dimgray', label= 'Largest telescopes in multiwavelengths')
plt.legend(loc = 'lower right',bbox_to_anchor=(1.01, 1.05),ncol=1)
m.plot(df_Radio['Longitude_East'],df_Radio['Latitude_North'],'*',color='dimgray') 
m.plot(df_Gamma['Longitude_East'],df_Gamma['Latitude_North'],'*',color='dimgray')
m.plot(df_Cos['Longitude_East'],df_Cos['Latitude_North'],'*',color='dimgray')
m.plot(df_N['Longitude_East'],df_N['Latitude_North'],'*',color='dimgray')
plt.savefig('Plots/All_Telescopes.pdf')
```
