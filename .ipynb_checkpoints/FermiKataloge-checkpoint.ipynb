```python
def get_Noff(alpha, bckgFactor, time, zd, energy_min, energy_max):
    off_rate, Noff = np.zeros(11), np.zeros(11)
        
    E = np.array([79.31, 125.89, 198.91, 315.99, 501.05, 794.38, 1258.925, 1995.39, 3162.42, 5011.78, 7943.81])   
    # search index for energy_min and energy_max 
    lowindex = np.argmax(E>energy_min)
    if energy_max > 7943:
        highindex = 9
    else:
        highindex = np.argmax(E>energy_max)
    print(lowindex,highindex)
    
    # bkg-rate from MAGIC paper. in 1/min
    if zd=='low':
        bkg_rate  = np.array([4.06,2.41,0.54,0.066,0.027,0.0133,0.0059,0.0027,0.0020,0.0014,0.0046])
    if zd=='mid':
        bkg_rate = np.array([2.92,2.89,0.54,0.305,0.088,0.038,0.0197,0.0111,0.0093,0.0042,0.0052])
        
    #for i in range(0,len(bkg_rate)): #sum of all 11 bins?? -> need depenence on energy_min + energy_max 
    for i in range(lowindex,highindex):
        # off_rate in 1/h
        off_rate[i] = (bkg_rate[i]/alpha) * 60.0 * bckgFactor
        # number off events in specific observation time
        Noff[i] = off_rate[i] * time

    return np.sum(Noff)
```

```python
def get_Aeff(energy, zd):
    E = np.array([79.31, 125.89, 198.91, 315.99, 501.05, 794.38, 1258.925, 1995.39, 3162.42, 5011.78, 7943.81])
    if zd=='low':
        Aeff = np.array([1.244,1.9732,3.067,2.497,4.494,6.505,8.591,8.909,9.888,11.094,14.35251])
    if zd=='mid':
        Aeff = np.array([0.16529681,1.46266036,2.43023741,5.25106048,6.48342453,10.64483946,13.92425647,19.89855896,
                  19.05061758,23.85890173,23.92085088]) 
        
    Aeff = Aeff*10**8 # cm^2
    # interpolate
    return np.interp(energy, E, Aeff, left=0)
```

```python
# calculate flux for given energy     

def calc_flux(source, energy):
    E0 = float(fermi.Pivot_Energy[fermi.Fermi_name == source]/1000.) # GeV
    K = float(fermi.Flux_Density[fermi.Fermi_name == source]*1000.) # 1. / cm^2 / GeV / s
    
    if fermi.SpectrumType[fermi.Fermi_name == source].all()=='LogParabola':
        alpha  = float(fermi.Spectral_Index[fermi.Fermi_name == source])
        beta = float(fermi.beta[fermi.Fermi_name == source])
        return K * (energy/E0)**(-alpha-beta*math.log(energy/E0))  
        
    if fermi.SpectrumType[fermi.Fermi_name == source].all()=='PowerLaw':    
        slope  = float(fermi.PowerLaw_Index[fermi.Fermi_name == source])
        return K * (energy/E0)**(-slope)
    
    if fermi.SpectrumType[fermi.Fermi_name == source].all()=='PLSuperExpCutoff':
        gamma  = float(fermi.Spectral_Index[fermi.Fermi_name == source])
        b = float(fermi.Exp_Index[fermi.Fermi_name == source])
        Ec = float(fermi.Cutoff[fermi.Fermi_name == source])/1000. # GeV
        return K * (energy/E0)**(-gamma)*math.exp((E0/Ec)**b - (energy/Ec)**b)    
```

```python
# EBL absorption

ebl = pd.read_csv('ebl2.csv',sep=' ',decimal=',', header=None)
ebl = ebl.astype(float) # converts entered values for Tau to floats 

arrayZ = np.array([0.01, 0.02526316, 0.04052632, 0.05578947, 0.07105263, 0.08631579, 0.10157895,
                   0.11684211 ,0.13210526,0.14736842,0.16263158,0.17789474,0.19315789,0.20842105,
                   0.22368421,0.23894737,0.25421053,0.26947368,0.28473684,0.3,0.35,0.4,0.45,0.5,0.55,
                   0.6,0.65,0.7,0.75,0.8,0.85,0.9,0.95,1.0, 1.2, 1.4, 1.6, 1.8, 2.0]) #Values for z 

Energy  = np.array([2.999999999999999889,3.454190000000000038
,3.977130000000000248,4.579249999999999987,5.272530000000000272,6.070770000000000333
,6.989860000000000517,8.048089999999999411,9.266530000000000600,10.66939999999999972,12.28469999999999979
,14.14459999999999884,16.28600000000000048,18.75169999999999892,21.59059999999999868,24.85930000000000084
,28.62290000000000112,32.95620000000000216,37.94569999999999887,43.69049999999999878,50.30499999999999972,
57.92089999999999739,66.68990000000000196,76.78639999999999910,88.41160000000000130,101.7970000000000041,117.2080000000000011
,134.9529999999999896,155.3840000000000110,178.9090000000000069,205.9950000000000170,237.1809999999999974,273.0890000000000040
,314.4340000000000135,362.0379999999999932,416.8490000000000251,479.9579999999999735,552.6209999999999845,636.2849999999999895
,732.6159999999999783,843.5309999999999420,971.2369999999999948,1118.280000000000030,1287.579999999999991,
1482.510000000000083,1706.960000000000122,1965.390000000000015,2262.940000000000040,2605.529999999999902,3000.000000000000000
]) # values for E 

Energy = Energy*10 # in GeV

def get_tau(z,energy):
    if  z>=2 or z<0.01 :
        Tau = 0 # This value range is not given inredshift range ######")
        print("#### no tau estimation possible!!!! Redshift out of range ######")
    if (energy<30) or (energy>=30000): # in GeV
        Tau = 0 
        print("#### no tau estimation possible!!!! Energy out of range ######")
    else:
        indexZ = np.argmax(arrayZ>z)
        zlow = arrayZ[indexZ-1] # z smaller than entered z 
        zhigh = arrayZ[indexZ] # z bigger than entered z 
        indexE=np.argmax(Energy>energy)
        elow = Energy[indexE-1]# Energy smaller than entered E
        ehigh = Energy[indexE] # Energy bigger than entered E 
        #  Start interpolation 
        ebl11 = ebl.iloc[indexE-1][indexZ-1]
        ebl12 = ebl.iloc[indexE-1][indexZ]
        ebl21 = ebl.iloc[indexE][indexZ-1]
        ebl22 = ebl.iloc[indexE][indexZ]
        zfrac = (z-zlow)/(zhigh-zlow)
        efrac = (energy-elow)/(ehigh-elow) 
        zinterpol1 = ebl11+zfrac*(ebl12-ebl11)
        zinterpol2 = ebl21+zfrac*(ebl22-ebl21)
        Tau = zinterpol1+efrac*(zinterpol2-zinterpol1)
        
    return Tau 

def get_absorption(source, energy):
    #
    z = float(fermi.Redshift[fermi.Fermi_name == source])
    z=0.02
    return math.exp(-get_tau(z,energy))

def get_absorpt(redshift,energy):
    z = redshift 
    return math.exp(-get_tau(z,energy))
```

```python
# calculate number excess events
def calc_excess_rate(source, energy_min, energy_max, zd):
    stepsize = 1. #GeV
    steps = int((energy_max-energy_min)/stepsize)
    #print(steps)
    rate = 0
    for step in range(0,steps):
        energy = energy_min + step * stepsize
        #print(energy)
        rate += stepsize * calc_flux(source, energy) * get_absorption(source, energy) * get_Aeff(energy, zd)
        #interpolated effective collection Area per energy 
    return rate
```

```python
def Get_Crab(x):
    f0 = 3.39*10**(-11)
    f0 = f0*10**(-3) # per GeV 
    aa = 2.51
    b = -0.21
    E0 = 1000 
    return f0*(x/E0)**((-aa)+b*np.log10(x/E0))
     #return f0*(x/E0)**((-alpha)+beta*np.log10(x/E0))
    
def plot_Crab_quad():
    x = np.linspace(10,10000,10000)
    plt.plot(x,Get_Crab(x)*x**2,'-',color='darkblue',label='$\mathrm{Crab Nebula}$',)
    
def plot_Crab_quad_less(percent):
    x = np.linspace(50,5000,1000)
    plt.plot(x,Get_Crab(x)*x**2*percent,'--',color='darkblue')
    
def plot_Crab():
    x = np.linspace(1,10000,10000)
    plt.plot(x,Get_Crab(x),'-',color='lightblue',label='$\mathrm{Crab}$',)
    
def plot_Crab_less(percent):
    x = np.linspace(50,5000,1000)
    plt.plot(x,Get_Crab(x)*percent,'--',color='lightsteelblue')
```

```python
def calc_flux2(energy):
    if spectrumtype== 'PowerLaw':
        return K * (energy/E0)**(-slope)
    if spectrumtype =='LogParabola':
        return K * (energy/E0)**(-alpha-beta*math.log(energy/E0))
    
def calc_excess_analyse(energy_min,energy_max,zd):
    stepsize = 1. #GeV
    steps = int((energy_max-energy_min)/stepsize)
    #print(steps)
    rate = 0
    for step in range(0,steps):
        energy = energy_min + step * stepsize
        #print(energy)
        rate += stepsize * calc_flux2(energy) * get_absorpt(redshift, energy) * get_Aeff(energy, zd)
        #interpolated effective collection Area per energy 
    return rate

def plot_point(Flux,Err_Flux):
    Flux = Flux/150.05    # From integral Flux to differential flux 
    ErrFlux = Err_Flux/150.05 
    PlotErrorBar = (Flux/f0 * (10)**(-PowerLawIndex))**(-1/PowerLawIndex)
    print(PlotErrorBar)
    plt.errorbar(PlotErrorBar,Flux,xerr=[[PlotErrorBar-0.1],[300-PlotErrorBar]],yerr=[ErrFlux], 
                fmt=None,color='crimson',ecolor='crimson',
             linewidth = 2, capsize=3, errorevery=1, label=r'$\textit{Fermi}$ $\rm{analysis}$ $\rm{measurement}$')
    x = np.linspace(0.1,50000)
    plt.plot(x,calc_flux2(x),'k--',label='$\mathrm{Extrapolated}$ $\mathrm{differential}$ $\mathrm{flux}$')
    x = np.linspace(0.1,29000)
    Abs = np.zeros(len(x))
    for i in range(0,len(x)): # Plot Flux* E^2
        Abs[i] = calc_flux2(x[i]) * get_absorpt(redshift, x[i])
    plt.plot(x,Abs,color='black',label='$\mathrm{Absorbed}$ $\mathrm{flux}$')
    plt.xscale('log')
    plt.yscale('log')
    plt.ylim(1*10**(-20))
    plt.tick_params(axis='x', labelsize=14)
    plt.tick_params(axis='y', labelsize=14)
    plt.xlabel(r'$\rm{E / GeV}$',fontsize=13)
    plt.ylabel(r'$\frac{\rm{d}\phi}{\rm{d E}$ / $\rm{cm}^{-2}$ $\rm{s}^{-1}$ $\rm{GeV}^{-1}$ ',fontsize=15)
    plt.legend(loc='upper right',fontsize=14)
    fig=plt.gcf()

    fig.set_size_inches(9,6)

    

def plot_point_quad(Flux,Err_Flux):
    Flux = Flux/150.05    # From integral Flux to differential flux 
    ErrFlux = Err_Flux/150.05 
    PlotErrorBar = (Flux/f0 * (10)**(-PowerLawIndex))**(-1/PowerLawIndex)
    print(PlotErrorBar)
    #plt.errorbar(PlotErrorBar,Flux*PlotErrorBar**2,xerr=[[PlotErrorBar-0.1],[300-PlotErrorBar]],yerr=[ErrFlux*PLotErrorBar**2], 
           #     fmt=None,color='crimson',ecolor='crimson',
           #  linewidth = 2, capsize=3, errorevery=1, label=r'$\textit{Fermi}$ $\rm{"quick}$-$\rm{look"}$ $\rm{Flusspunkt}$')
    x = np.linspace(0.1,50000)
    plt.plot(x,calc_flux2(x)*x**2,'k--',label='$\mathrm{Extrapolated}$ $\mathrm{differential}$ $\mathrm{flux}$')
    x = np.linspace(0.1,29000)
    Abs = np.zeros(len(x))
    for i in range(0,len(x)): # Plot Flux* E^2
        Abs[i] = calc_flux2(x[i]) * get_absorpt(redshift, x[i])
    plt.plot(x,Abs*x**2,color='black',label='$\mathrm{Absorbed}$ $\mathrm{flux}$')
    plt.xscale('log')
    plt.yscale('log')
    plt.ylim(10**(-11))
    plt.tick_params(axis='x', labelsize=14)
    plt.tick_params(axis='y', labelsize=14)
    plt.xlabel(r'$\rm{E / GeV}$',fontsize=13)
    plt.ylabel(r'$\frac{\rm{d}\phi}{\rm{d E}$ / $\rm{cm}^{-2}$ $\rm{s}^{-1}$ $\rm{GeV}^{-1}$ ',fontsize=15)
    plt.legend(loc='upper right',fontsize=14)
    fig=plt.gcf()

    fig.set_size_inches(9,6)


    
```
