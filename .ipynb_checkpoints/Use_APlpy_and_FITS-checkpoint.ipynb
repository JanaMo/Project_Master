## Anwendung auf ctool Ergebnisse 

Ctools gibt uns nach **ctobssim** eine *.fits* mit einem leeren PRIMARY, den EVENTs, in den Events, wobei jede zeile = ein Event (=Photon) und die GTI matrix mit Good Timing Intervals <br>
Events beinhaltet:
- EVENT_ID 
- TIME
- RA / DEC
- ENERGY
- DETX
- DETY
- MC_ID

und zwar so h√§ufig, wie Photonen auftreten 

## nach Eventbinning von ctools 

```python
Datei = fits.open('Aplpy/cntcube.fits',ignore_missing_end=True)
Datei.info()
Counts3D = Datei['COUNTS']
print(len(Counts3D.data[0][0]))
Energy1 = Counts3D.data[0]
        
plt.imshow(Energy1,cmap='gist_gray')
```

## Verpacken von Fits Dateien, sodass die von Aplpy genutzt werden k√∂nnen --> Ein Image File erzeugen 

```python
#################################################################
############################# Header ############################
#################################################################
Head = fits.Header()
Head['Built'] = 'Ctools'
Head['NAXIS']   =                    2                                               
Head['NAXIS1']  =                  200                                                  
Head['NAXIS2']  =                  200                                                  
                                                                                                         
Head['CDATE']   = 'Wed Feb 25 11:57:21 2009'                                            
Head['CTYPE1']  = 'RA---TAN'                                                            
Head['CTYPE2']  = 'DEC--TAN'                                                            
Head['CRVAL1']  =   83.6                                                  
Head['CRVAL2']  =   22.1                                                 
Head['CRPIX1']  =   361.                                                  
Head['CRPIX2']  =   360.5                                                                                                                                               
Head['EQUINOX'] =   2000.0                                                     
##################################################################
########################## HDULIST ###############################
##################################################################
hdu = fits.PrimaryHDU(Energy1,header=Head)
hdulist = fits.HDUList([hdu])
hdulist.writeto('Aplpy/new.fits')


####################################################################
################################## Test ############################
####################################################################
Datei = fits.open('Aplpy/new.fits',ignore_missing_end=True)
Datei.info()
gc = aplpy.FITSFigure('Aplpy/new.fits')
gc.show_grayscale()
```
